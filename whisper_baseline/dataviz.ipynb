```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('results.csv')
df = df[df['file_missing?']==False]
```

```python
alt.Chart(df).mark_point().encode(
    x='mean_wer:Q',
    y='native_language',
    tooltip=['mean_wer:Q']
).transform_aggregate(
    mean_wer='mean(wer)',
    groupby=["native_language"]
)
```

```python
alt.Chart(df).mark_point().encode(
    x='mean_wil:Q',
    y='native_language',
    tooltip=['mean_wil:Q']
).transform_aggregate(
    mean_wil='mean(wil)',
    groupby=["native_language"]
)
```
